```python
import pyodbc
import pandas as pd

# Thông tin kết nối
server = r'MSI\SQLEXPRESS'   # tên server instance trong SSMS
database = 'DB_ECommerce'        # tên database bạn muốn truy vấn

# Tạo kết nối với Windows Authentication
conn = pyodbc.connect(
    'DRIVER={ODBC Driver 17 for SQL Server};'
    f'SERVER={server};'
    f'DATABASE={database};'
    'Trusted_Connection=yes;'
)

# Viết câu lệnh SQL
query = "SELECT * FROM dbo.olist_seller_timestamp"

# Đọc dữ liệu vào DataFrame
df = pd.read_sql(query, conn)

# Xuất ra file CSV
df.to_csv("olist_seller_timestamp.csv", index=False, encoding="utf-8")

print("Dữ liệu đã được lưu vào olist_seller_timestamp.csv")

# Đóng kết nối
conn.close()
```
